```python
tahun = ['NULL','0015','0016','0017','0020','0021','0022','0028','0029','0200','0201','0202','0207','0214','0220','0223','1010','1977',
         '1985','1990','1991','1992','1992','1993','1994','1995','1996','1997','1998','1999','2000','2001','2002','2003','2004','2005',
         '2006','2007','2008','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019','2020','2021','2022','2023',
         '2107','3018']
links = []

for i in range(len(tahun)):
  variabel = str(tahun[i])
  link = "http://repository.lppm.unila.ac.id/cgi/exportview/year/"+variabel+"/JSON/"+variabel+".js"
  links.append(link)
  print(links[i])
```

##Buat dataset dari repository

Masing-masing repository digabung jadi 1 dataset besar dengan nama "hasil"

```python
# df = []
# df.append(pd.read_json("http://repository.lppm.unila.ac.id/cgi/exportview/year/3018/JSON/3018.js"))
# df.append(pd.read_json("http://repository.lppm.unila.ac.id/cgi/exportview/year/2107/JSON/2107.js"))
# hasil = pd.concat(df)
# hasil

df = []
for i in range(len(links)):
  df.append(pd.read_json(links[i]))

hasil = pd.concat(df)
```

##Mengambil kolom yang diperlukan
Data yang perlu di ambil meliputi division, title, abstract, subjects, publication, publisher, date, keywords

```python
hasil_df = hasil[['divisions','title','abstract','subjects','publication','publisher','date','keywords']]
```
