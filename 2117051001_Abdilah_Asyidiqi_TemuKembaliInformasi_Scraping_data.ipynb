##Bikin list berisi link ke repository JSON

```python
url = requests.get("http://repository.lppm.unila.ac.id/view/year/").text

soup = bs4(url,"lxml")
scrap = soup.find_all('li' > 'a href', {"class":"ep_view_cols ep_view_cols_3"} )
```

```python
list_tahun = []

for ul in scrap:
    li_data = ul.find_all('li')
    for li in li_data:
        a = li.find('a')
        hasil = a.get('href')
        x = hasil.replace(".html","")
        list_tahun.append(x)

print(list_tahun)
```

```python
links = []

for i in range(len(list_tahun)):
  variabel = str(list_tahun[i])
  link = "http://repository.lppm.unila.ac.id/cgi/exportview/year/"+variabel+"/JSON/"+variabel+".js"
  links.append(link)
  print(links[i])
```

##Buat dataset dari repository

Masing-masing repository digabung jadi 1 dataset besar dengan nama "hasil"

```python
df = []
for i in range(len(links)):
  df.append(pd.read_json(links[i]))

hasil = pd.concat(df)
```
