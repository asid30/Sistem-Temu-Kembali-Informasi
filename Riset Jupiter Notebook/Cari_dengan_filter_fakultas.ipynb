```python
new_divisions = []
for i in range(len(df['divisions'])):
  try:
    new_divisions.append(ast.literal_eval(df['divisions'].iloc[i]))
  except:
    new_divisions.append(ast.literal_eval('[]'))
```

```python
df['new_divisions'] = new_divisions
```

```python
import re

def pisahkan_kata(kata):
    hasil = re.sub(r'\d', '', kata)
    return hasil
```

```python
new_fakultas = []
for i in range(len(df)):
  temp = df['new_divisions'].iloc[i]
  temp2 = []
  for j in range(len(temp)):
    temp2.append(pisahkan_kata(temp[j]))
  new_fakultas.append(temp2)
```

```python
new_fakultas
```

```python
df['pra_fakultas'] = new_fakultas
```

```python
df[['divisions','title','new_divisions','pra_fakultas']]
```

```python
df_hasil = []
for i in range(len(df)):
  if 'FK' in df['pra_fakultas'].iloc[i]:
    df_hasil.append(df.iloc[i])
```
